**Visualizations**

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df,x='Review Rating',kde=True)
plt.xlabel('Review Rating Distribution',fontsize = 15,weight='bold')
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'] , kde=True, color = 'purple')
plt.title('Age Distribution', fontsize=15, weight='bold')
plt.xlabel('Age',fontsize = 15,weight='bold')
plt.show()
```

```python
sns.countplot(data=df,x='Category',palette='Set2')
plt.title('Category Distribution', fontsize=15, weight='bold')
#plt.xlabel('Category',fontsize = 15,weight='bold')
plt.show()
```

```python
sns.countplot(data=df,x='Payment Method',palette='Set1')
plt.title('Payment Methods', fontsize=15, weight='bold')
plt.show()
```

```python
plt.pie(df['Gender'].value_counts(), labels=df['Gender'].unique(), autopct='%1.1f%%',colors=['blue','pink'])
plt.title('Gender Distribution', weight='bold',fontsize=15)
plt.show()
```

```python
top = df['Item Purchased'].value_counts().sort_values(ascending=False).head(10)
sns.barplot(x=top.index, y=top.values,palette='rainbow')
plt.xlabel('Item')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.title('Top 10 Most Frequently Purchased Items', weight='bold',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Gender', y='Purchase Amount (USD)', estimator='mean', errorbar=None, hue='Gender', palette=['blue','pink'], legend=False)
plt.title('Average Purchase Amount by Gender')
plt.ylabel('Average Purchase Amount (USD)')
plt.xlabel('Gender')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Category', y='Purchase Amount (USD)', estimator='mean', errorbar=None, hue='Category')
plt.title('Average Purchase Amount by Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.xlabel('Category')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Season', y='Purchase Amount (USD)', estimator='mean', errorbar=None, hue='Season', palette='Set2')
plt.title('Average Purchase Amount by Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.xlabel('Season')
plt.show()
```

```python
customer_location = df['Location'].value_counts(ascending=False).head(10)
sns.barplot(x=customer_location.index, y=customer_location.values, palette="Set1")
plt.xticks(rotation=45)
plt.ylabel("Number of Customers")
plt.title("Locations Having Highest Number of Customer", weight='bold',fontsize=15)
plt.show()
```

```python
highest_revenue = df.groupby("Category")["Purchase Amount (USD)"].sum().sort_values(ascending=False)
sns.barplot(x=highest_revenue.index, y=highest_revenue, palette="viridis")
plt.title("Category generating the highest total revenue", weight='bold',fontsize=15)
plt.show()
```

```python
plt.pie(df['Subscription Status'].value_counts(),labels=['Subscription','no Subscription'],autopct='%1.1f%%', colors=['Green','Teal'])
plt.title('Subscription Status', weight='bold',fontsize=15)
plt.show()
```

```python
discounts= df[df['Discount Applied']=="Yes"]
most_purchased_item = discounts['Item Purchased'].value_counts(ascending=False).head(10)


sns.barplot(x=most_purchased_item.index, y = most_purchased_item.values, palette="rainbow")
plt.title("Most purchased when discounts are applied", weight='bold',fontsize=15)
plt.xticks(rotation = 45)
plt.ylabel("No. of Items")
plt.show()
```

```python

most_common_shipping = df['Shipping Type'].value_counts().idxmax()
shipping_counts = df['Shipping Type'].value_counts()

print("Most common shipping type:", most_common_shipping)
print("\nAll shipping type counts:")
print(shipping_counts)
```

```python
shipping_counts = df['Shipping Type'].value_counts().reset_index()
shipping_counts.columns = ['Shipping Type', 'Count']


plt.figure(figsize=(6, 4))
sns.barplot(data=shipping_counts, x='Shipping Type', y='Count', palette='Blues_d')
plt.title('Most Common Shipping Types')
plt.xlabel('Shipping Type')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
